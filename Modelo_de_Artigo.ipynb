A revista Collaborative Paper tem como missão promover a excelência acadêmica e a inovação no ensino da Matemática e disciplinas relacionadas. Nosso foco é na publicação de trabalhos que sejam colaborativos e interativos. Estamos comprometidos em criar um ambiente de aprendizado dinâmico e inclusivo no Instituto Federal da Bahia, proporcionando um espaço onde estudantes e professores possam compartilhar conhecimento, descobrir novas ideias e fortalecer nossa comunidade acadêmica. Nossa revista mantém um alto padrão de qualidade, fomentando a pesquisa, o pensamento crítico e a criatividade, e simultaneamente integrando tecnologias inovadoras nas práticas educacionais. A Collaborative Paper não só destaca o trabalho excepcional de nossos alunos, mas também se estabelece como um recurso valioso para futuras gerações de estudiosos e apaixonados pela matemática e áreas relacionadas.

#### <font color='red'> **Importante**</font>

Os artigos publicados nesta plataforma são fruto dos projetos realizados por estudantes nas disciplinas que ministro no Instituto Federal da Bahia, campus Vitória da Conquista. No caso de um artigo ser selecionado para publicação em outro veículo, ele será removido daqui, contanto que recebamos uma solicitação formal nesse sentido. Ademais, os trabalhos podem ser retirados a qualquer momento a pedido de quaisquer envolvidos, incluindo docentes, discentes ou outros integrantes do instituto. Com a anuência dos alunos, minha participação nos projetos pode ir além da mera formatação para esta revista. Além disso, outros professores e especialistas podem ser convidados a contribuir para o enriquecimento dos trabalhos, sempre com a concordância de todas as partes envolvidas.
